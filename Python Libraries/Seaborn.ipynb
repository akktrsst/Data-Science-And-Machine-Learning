## Visualizing Statistical Relationships

```python
df = sns.load_dataset('tips') #load 'tips' dataset from python library
df.head()
```

```python
sns.relplot(x='total_bill', y= 'tip', data= df) # this way of plotting can't be done in matplotlib
plt.show()
```

```python
sns.relplot(x='total_bill', y= 'tip', data= df, kind='line') # this way of plotting can't be done in matplotlib
# kind is showing type of graph
plt.show()
```

```python
sns.relplot(x='total_bill', y= 'tip', data= df, kind='scatter') # this way of plotting can't be done in matplotlib
# kind is showing type of graph
plt.show()
```

```python
sns.relplot(x='total_bill', y= 'tip', data= df, hue='time') 
# based on time we categorized data by using parameter "hue='time'" =color
plt.show()
```

```python
sns.relplot(x='total_bill', y= 'tip', data= df, style='time') 
# based on time we categorized data by using parameter "style='time'" =shape of scatter
plt.show()
```

```python
sns.relplot(x='total_bill', y= 'tip', data= df, style='time', hue= 'smoker') 
plt.show()
```

```python
sns.relplot(x='total_bill', y= 'tip', data= df, hue='time', col= 'smoker') 
# col= it will divide in two part
plt.show()
```

```python
sns.scatterplot(x='total_bill', y= 'tip', data= df, hue='time') 
# scatterplot is similiar except some parameter, like= col, we can't use this parameter in scatterplot method
plt.show()
```

## Plotting with Categorical Data

## Categorical Scatterplots

```python
Syntax:-
seaborn.catplot(x=value, y=value, data=data)

The default representation of data in catplot() uses a scatterplot(stripplot).
```

```python
df= sns.load_dataset('tips')
df.head()
```

```python
sns.catplot(x='sex',y='total_bill', data=df)
plt.show()
```

```python
sns.catplot(x='sex',y='total_bill', data=df, jitter=1)
plt.show()
```

```python
sns.catplot(x='sex',y='total_bill', data=df, jitter=0.7) 
# jitter value gives the randomness of scatter, 
plt.show()
```

```python
sns.catplot(x='sex',y='total_bill', data=df, jitter=0.005) 
# jitter value gives the randomness of scatter, 
plt.show()
```

```python
sns.catplot(x='sex',y='total_bill', data=df, kind='swarm') 
# not overlapping
plt.show()
```

```python
sns.catplot(x='sex',y='total_bill', data=df, jitter=0.005, hue='smoker') 
# overlapping
plt.show()
```
